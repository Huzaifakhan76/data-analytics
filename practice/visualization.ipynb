```python
import matplotlib.pyplot as plt
x = ['mango', 'banana', 'graphes', 'apple']
y = [70,20,100,66]
plt.xlabel("fruits")
plt.ylabel('number of usage')
c = ("y",'y','g','r')
plt.bar(x,y,color = c, alpha = 0.5,linestyle = ":", label = "FRUITS"  )
plt.legend()
plt.show()
```

```python
x = [90,7,22,5,90,13,17,4]
y = [7,2,7,66,70,88,90,22]
size = [100,200,400,500,200,350,330,200]
c = ("y","g","w","b","g","r","g",'b')
plt.scatter(x,y, label="numbers list", s = size , color= c, edgecolors= "g" , linewidths= 1)
plt.colorbar()

plt.ylabel("percentage")
plt.xlabel("numbers")
plt.legend()

plt.show()
```
